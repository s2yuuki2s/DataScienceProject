```python
# Đọc dữ liệu từ file gốc
file_path = '../data/raw/product_details.csv'  # Thay thế bằng tệp dữ liệu ban đầu của bạn
data = pd.read_csv(file_path)

# Tiền xử lý dữ liệu
# 1. Loại bỏ cột không cần thiết
data = data.drop(columns=['store_name', 'link'])

# 2. Loại bỏ các dòng trùng lặp
data = data.drop_duplicates()

# 3. Xử lý giá trị thiếu
data = data.dropna()

# 4. Xử lý các cột số liệu
data['comment_count'] = data['comment_count'].str.extract(r'(\d+)').fillna(0).astype(int)
data['price'] = data['price'].str.replace(r'[^\d]', '', regex=True).astype(int)

# Xử lý cột 'sold' bao gồm các giá trị như '.k'
def convert_sold_value(value):
    if isinstance(value, str):
        value = value.lower().replace('sold', '').strip()  # Loại bỏ từ 'sold'
        if 'k' in value:
            return int(float(value.replace('k', '')) * 1000)
        elif 'm' in value:
            return int(float(value.replace('m', '')) * 1000000)
        else:
            return int(value)
    return 0

data['sold'] = data['sold'].apply(convert_sold_value)

data['rating'] = pd.to_numeric(data['rating'], errors='coerce')

# Lưu kết quả vào file preprocessing.csv
preprocessing_file_path = '../data/processed/preprocessing.csv'
data.to_csv(preprocessing_file_path, index=False)

print("Hoàn thành tiền xử lý và lưu vào ../data/processed/preprocessing.csv")
data.head()
```
