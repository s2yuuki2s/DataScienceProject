```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import pandas as pd
```

```python
#train data
train_data = pd.read_csv("../data/processed/train_data.csv")
train_data.info()
train_data.describe()
train_data.head()
```

```python
#test data 
test_data = pd.read_csv("../data/processed/test_data.csv")
test_data.info()
test_data.describe()
test_data.head()
```

```python
# Chọn các biến độc lập và mục tiêu
X_train = train_data[['rating', 'comment_count']] 
y_train = train_data['sold']  

X_test = test_data[['rating', 'comment_count']]  
y_test = test_data['sold']  

# Huấn luyện mô hình Ridge Regression
model = Ridge(alpha=1.0)  # alpha là tham số điều chỉnh độ mạnh của regularization
model.fit(X_train, y_train)

# Dự đoán trên dữ liệu kiểm tra
y_pred = model.predict(X_test)

# Đánh giá mô hình
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\nĐánh giá mô hình:")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")

# Hiển thị kết quả dự đoán
results = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print("\nKết quả dự đoán (mẫu):")
print(results.head())
```
